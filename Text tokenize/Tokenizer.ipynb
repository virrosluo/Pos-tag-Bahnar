```python
corpus = []
with open('Text_dataset\\Vietnam_0.txt', 'r', encoding='utf-8') as fp:
    for i in range(23):
        str = fp.readline()
        corpus.append(str.split())

print(corpus[1])
```

    GENERATE ALL TOKEN

```python
def token_generation(sentence, dict):
    total_count = 0
    for i in range(1, len(sentence)):
        for j in range(len(sentence)):
            if j + i <= len(sentence):
                str = ' '.join(sentence[j: j + i])
                if str in dict:
                    dict[str] += 1
                else:
                    dict[str] = 1
                
                total_count += 1
    return dict, total_count
```

```python
dict = {}
total_count = 0
for sentence in corpus:
    dict, temp_count = token_generation(sentence, dict)
    total_count += temp_count
```
